```python

# Load CSV files
gpt35 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Raffaele/1/01_Aggression_NEW_with_answers_GPT35.csv')

gpt4 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Raffaele/1/01_Aggression_NEW_with_answers_GPT4.csv')
```

```python
# Define a mapping for your annotations
annotation_mapping = {'aggressive': 1, 'not aggressive': 0}

```

```python
results_df = pd.DataFrame(columns=['Run', 'F1_Macro_gpt4', 'Accuracy_gpt4'])

answer_columns = ['answerPrompt2', 'answerPrompt4', 'answerPrompt6', 'answerPrompt8', 'answerPrompt10', 'answerOLD']

print(gpt4.shape)

rows_to_remove = [136, 147, 170, 175, 237, 498, 981, 982, 983, 985, 986, 990, 991]
gpt4_filtered = gpt4.drop(rows_to_remove)

#gpt4_filtered=gpt4[gpt4['answerOLD'] != 'None']

print(gpt4_filtered.shape)

y_true = gpt4_filtered['annotation'].map(annotation_mapping)

# Iterate over each run configuration
for run_config in answer_columns:

    y_pred = gpt4_filtered[run_config].map(annotation_mapping)  # Predictions for the current run configuration

    # Check for NaN values in y_true and y_pred
    nan_positions_true = y_true.isna().any()
    nan_positions_pred = y_pred.isna().any()

    if nan_positions_true or nan_positions_pred:
        print(f"NaN values found in sheet run configuration '{run_config}':")
        print("NaN positions in y_true:")
        print(y_true[y_true.isna()])
        print("\nNaN positions in y_pred:")
        print(gpt4_filtered[run_config][y_pred.isna()])


    conf_matrix = confusion_matrix(y_true, y_pred)

    # Calculate accuracy
    accuracyGPT4 = accuracy_score(y_true, y_pred)

    # Calculate F1 score macro
    f1GPT4= f1_score(y_true, y_pred, average='macro')

    new_row = pd.DataFrame([{
        'Run': run_config,
        'F1_Macro_gpt4': f1GPT4,
        'Accuracy_gpt4': accuracyGPT4
    }])

    results_df = pd.concat([results_df, new_row], ignore_index=True)


print(results_df)

# Save results to a CSV file
results_df.to_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Raffaele/1/01_Aggressive_resultEvaluation_FILTERED.csv', index=False)
```
