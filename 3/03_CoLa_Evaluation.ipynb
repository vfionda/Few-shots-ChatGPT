```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score
import os
from google.colab import drive
```

```python
# Load CSV files
gpt35 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Raffaele/3/03_CoLa_NEW_with_answers_GPT35.csv')

gpt4 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Raffaele/3/03_CoLa_NEW_with_answers_GPT4.csv')
```

```python
results_df = pd.DataFrame(columns=['Run', 'F1_Macro_gpt35', 'Accuracy_gpt35', 'F1_Macro_gpt4', 'Accuracy_gpt4'])

answer_columns = ['answerPrompt2', 'answerPrompt4', 'answerPrompt6', 'answerPrompt8', 'answerPrompt10', 'answerOLD']
answer_columns2 = ['answerPrompt2', 'answerPrompt4', 'answerPrompt6', 'answerPrompt8', 'answerPrompt10']


y_true = gpt35['binaryvalue'].astype(int)


# Iterate over each run configuration

for column_name in answer_columns2:
  try:
    # Try to convert the column to integer
    gpt35[column_name] = gpt35[column_name].astype(int)
  except ValueError as e:
    # If conversion to integer fails, print the row causing the error
    error_row_index = gpt35[gpt35[column_name].apply(lambda x: not x.isdigit())].index[0]
    error_row = gpt35.loc[error_row_index]
    print(f"Row causing the error:\n{error_row}")

for column_name in answer_columns:
  try:
    # Try to convert the column to integer
    gpt4[column_name] = gpt4[column_name].astype(int)
  except ValueError as e:
    # If conversion to integer fails, print the row causing the error
    error_row_index = gpt4[gpt4[column_name].apply(lambda x: not x.isdigit())].index[0]
    error_row = gpt4.loc[error_row_index]
    print(f"Row causing the error:\n{error_row}")


for run_config in answer_columns:
  print (run_config)
  if run_config!='answerOLD':
    y_pred = gpt35[run_config].astype(int)  # Predictions for the current run configuration

    # Calculate accuracy
    accuracyGPT35 = accuracy_score(y_true, y_pred)

    # Calculate F1 score
    f1GPT35 = f1_score(y_true, y_pred)


  else:
    accuracyGPT35=0
    f1GPT35=0

  y_pred = gpt4[run_config].astype(int)  # Predictions for the current run configuration

   # Calculate accuracy
  accuracyGPT4 = accuracy_score(y_true, y_pred)

  # Calculate F1 score
  f1GPT4 = f1_score(y_true, y_pred)

  new_row = pd.DataFrame([{
        'Run': run_config,
        'F1_Macro_gpt35': f1GPT35,
        'Accuracy_gpt35': accuracyGPT35,
        'F1_Macro_gpt4': f1GPT4,
        'Accuracy_gpt4': accuracyGPT4
  }])

  results_df = pd.concat([results_df, new_row], ignore_index=True)


print(results_df)

results_df.to_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Raffaele/3/03_CoLa_resultEvaluation.csv', index=False)
```
