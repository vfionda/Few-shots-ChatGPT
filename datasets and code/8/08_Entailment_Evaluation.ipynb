```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import os
from google.colab import drive
from sklearn.metrics import classification_report
```

```python
# Load CSV files
gpt35 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Sebastiano/8/08_Entailment_with_answers_GPT35.csv')

gpt4 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Sebastiano/8/08_Entailment_with_answers_GPT4.csv')
```

```python
annotation_mapping = {'entailment':0, 'not_entailment':1}
```

```python
results_df = pd.DataFrame(columns=['Run', 'F1_Macro_gpt35', 'Accuracy_gpt35', 'F1_Macro_gpt4', 'Accuracy_gpt4'])

y_true = gpt35['label'].map(annotation_mapping)

for run_config in ['answerPrompt2', 'answerPrompt4', 'answerPrompt6', 'answerPrompt8', 'answerPrompt10', 'answerOLD']:


    if run_config!='answerOLD':
      y_pred = gpt35[run_config].map(annotation_mapping)  # Predictions for the current run configuration


     # Check for NaN values in y_true and y_pred
      nan_positions_true = y_true.isna().any()
      nan_positions_pred = y_pred.isna().any()

      if nan_positions_true or nan_positions_pred:
        print(f"NaN values found in sheet run configuration '{run_config}':")
        print("NaN positions in y_true:")
        print(y_true[y_true.isna()])
        print("\nNaN positions in y_pred:")
        print(y_pred[y_pred.isna()])

      conf_matrix = confusion_matrix(y_true, y_pred)

      accuracyGPT35 = accuracy_score(y_true, y_pred)

      f1GPT35 = f1_score(y_true, y_pred, average='macro')
    else:
      accuracyGPT35 = 0
      f1GPT35 = 0

    y_pred = gpt4[run_config].map(annotation_mapping)

    # Check for NaN values in y_true and y_pred
    nan_positions_pred = y_pred.isna().any()

    if nan_positions_pred:
        print(f"NaN values found in sheet run configuration '{run_config}':")
        print("NaN positions in y_true:")
        print(y_true[y_true.isna()])
        print("\nNaN positions in y_pred:")
        print(y_pred[y_pred.isna()])

    conf_matrix = confusion_matrix(y_true, y_pred)

    accuracyGPT4 = accuracy_score(y_true, y_pred)

    f1GPT4 = f1_score(y_true, y_pred, average='macro')

    report_lines = classification_report(y_true=y_true, y_pred=y_pred, digits=4).split('\n')

    for line in report_lines:
      print(line)

    results_df = results_df.append({
        'Run': run_config,
        'F1_Macro_gpt35': f1GPT35,
        'Accuracy_gpt35': accuracyGPT35,
        'F1_Macro_gpt4': f1GPT4,
        'Accuracy_gpt4': accuracyGPT4
    }, ignore_index=True)

print(results_df)

results_df.to_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Sebastiano/8/08_Entailment_resultEvaluation.csv', index=False)
```
