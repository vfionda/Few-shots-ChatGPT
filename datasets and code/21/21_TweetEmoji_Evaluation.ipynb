```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score
import os
from google.colab import drive
import numpy as np
import ast
```

```python


# Load CSV files
gpt35 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Valeria/21/21_emoji_prompts_few_shots_with_answers_GPT3.5.csv')

gpt4 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Valeria/21/21_emoji_prompts_few_shots_with_answers_GPT4.csv')
```

```python
results_df = pd.DataFrame(columns=['Run', 'F1_Macro_gpt4','F1_Micro_gpt4', 'Accuracy_TOP1_gpt4', 'Accuracy_TOP3_gpt4'])

answer_columns = ['answerPrompt2', 'answerPrompt4', 'answerPrompt6', 'answerPrompt8', 'answerPrompt10', 'answerOLD']

gpt4_filtered=gpt4[gpt4.answerOLD != '[-1]']

y_true = gpt4_filtered['label'].to_numpy()


for system in answer_columns:

  gpt4_filtered[system+"_final"] = gpt4_filtered[system].apply(lambda x: ast.literal_eval(x))

  y_pred = gpt4_filtered[system].to_numpy()  # Predictions for the current run configuration

  gpt4_answer = gpt4_filtered[system+"_final"].to_list()
  y_pred_top1 = np.array([answer[0] for answer in gpt4_answer])
  y_pred_top3 = np.array(
  [
      y_true[i] if y_true[i] in answer else answer[0]
      for i, answer in enumerate(gpt4_answer)
  ]
  )

  f1_macro_gpt4 = f1_score(y_true=y_true, y_pred=y_pred_top1, average="macro")
  f1_micro_gpt4 = f1_score(y_true=y_true, y_pred=y_pred_top1, average="micro")
  accuracy_top1_gpt4 = accuracy_score(y_true=y_true, y_pred=y_pred_top1)
  accuracy_top3_gpt4 = accuracy_score(y_true=y_true, y_pred=y_pred_top3)

  results_df = results_df.append({
    'Run': system,
    'F1_Macro_gpt4': f1_macro_gpt4,
    'F1_Micro_gpt4': f1_micro_gpt4,
    'Accuracy_TOP1_gpt4': accuracy_top1_gpt4,
    'Accuracy_TOP3_gpt4': accuracy_top3_gpt4
  }, ignore_index=True)

print(results_df)


# Save results to a CSV file
results_df.to_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Valeria/21/21_emoji_EvaluationResults_FILTERED.csv', index=False)
```
