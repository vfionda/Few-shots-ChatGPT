```python


# Load CSV files
gpt35 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Valeria/18/18_uc_prompts_few_shots_with_answers_GPT3.5.csv')

gpt4 = pd.read_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Valeria/18/18_uc_prompts_few_shots_with_answers_GPT4.csv')
```

```python
annotation_mapping = {'antagonize':0, 'condescending':1, 'dismissive':2, 'generalisation':3, 'generalisation_unfair':4, 'healthy':5, 'hostile':6, 'sarcastic':7}


def map_annotation(annotation_list):
      return [annotation_mapping[string] for string in annotation_list]
```

```python

gpt35['annotation_int'] = gpt35['annotation'].apply(lambda x: map_annotation(ast.literal_eval(x)))
gpt4['annotation_int'] = gpt4['annotation'].apply(lambda x: map_annotation(ast.literal_eval(x)))


answer_columns35 = ['answerPrompt2', 'answerPrompt4', 'answerPrompt6', 'answerPrompt8', 'answerPrompt10']
answer_columns4 = ['answerPrompt2', 'answerPrompt4', 'answerPrompt6', 'answerPrompt8', 'answerPrompt10', 'answerOLD']

for system in answer_columns35:
  gpt35[system+'_int'] = gpt35[system].apply(lambda x: map_annotation(ast.literal_eval(x)))

for system in answer_columns4:
  gpt4[system+'_int'] = gpt4[system].apply(lambda x: map_annotation(ast.literal_eval(x)))
```

```python
# Iterate through answer columns
answer_columns = ['answerPrompt2_int', 'answerPrompt4_int', 'answerPrompt6_int', 'answerPrompt8_int', 'answerPrompt10_int', 'answerOLD_int']

results_df = pd.DataFrame(columns=['Run', 'F1_Macro_gpt4', 'Accuracy_gpt4'])

gpt4_filtered=gpt4[gpt4.answerOLD != '[]']

#
mlb = MultiLabelBinarizer()

for system in answer_columns:
    true_labels = mlb.fit_transform(gpt4_filtered['annotation_int'])


    # Compute F1 Macro and Accuracy for gpt4
    true_labels = mlb.fit_transform(gpt4_filtered['annotation_int'])
    predicted_labels = mlb.transform(gpt4_filtered[system])
    f1_macro_gpt4 = f1_score(true_labels, predicted_labels, average='macro')
    accuracy_gpt4 = accuracy_score(true_labels, predicted_labels)

    # Append results to the DataFrame
    results_df = results_df.append({
        'Run': system,
        'F1_Macro_gpt4': f1_macro_gpt4,
        'Accuracy_gpt4': accuracy_gpt4
    }, ignore_index=True)

print(results_df)

# Save results to a CSV file
results_df.to_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Valeria/18/18_uc_results_FILTERED.csv', index=False)
```

```python

# Iterate through answer columns
answer_columns = ['answerPrompt2_int', 'answerPrompt4_int', 'answerPrompt6_int', 'answerPrompt8_int', 'answerPrompt10_int', 'answerOLD_int']

results_df = pd.DataFrame(columns=['Run', 'F1_Macro_gpt35', 'Accuracy_gpt35', 'F1_Macro_gpt4', 'Accuracy_gpt4'])
#
mlb = MultiLabelBinarizer()

for system in answer_columns:
    true_labels = mlb.fit_transform(gpt35['annotation_int'])

    if system!='answerOLD_int':
      predicted_labels = mlb.transform(gpt35[system])
      f1_macro_gpt35 = f1_score(true_labels, predicted_labels, average='macro')
      accuracy_gpt35 = accuracy_score(true_labels, predicted_labels)
    else:
      f1_macro_gpt35 = 0
      accuracy_gpt35 = 0

    # Compute F1 Macro and Accuracy for gpt4
    true_labels = mlb.fit_transform(gpt4['annotation_int'])
    predicted_labels = mlb.transform(gpt4[system])
    f1_macro_gpt4 = f1_score(true_labels, predicted_labels, average='macro')
    accuracy_gpt4 = accuracy_score(true_labels, predicted_labels)

    new_row = pd.DataFrame([{
        'Run': system,
        'F1_Macro_gpt35': f1_macro_gpt35,
        'Accuracy_gpt35': accuracy_gpt35,
        'F1_Macro_gpt4': f1_macro_gpt4,
        'Accuracy_gpt4': accuracy_gpt4
    }])

    results_df = pd.concat([results_df, new_row], ignore_index=True)

print(results_df)

# Save results to a CSV file
results_df.to_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Valeria/18/18_uc_results.csv', index=False)

```

```python
######## F1 Macro per class, AVG wrt all classes########
def custom_accuracy_per_class(y_true, y_pred):
    # Calculate accuracy per class
    accuracies = []
    for class_index in list(annotation_mapping.values()):  # Iterate over each column/class
        # Custom accuracy calculation: True if the label is correctly predicted
        correct_predictions = np.sum((y_true[:, class_index] == y_pred[:, class_index]) | (y_true[:, class_index] == 0))
        accuracies.append(correct_predictions / y_true.shape[0])
    return accuracies

# Define custom F1-score per class
def custom_f1_score_per_class(y_true, y_pred):
    # Calculate F1 score per class
    f1_scores = []
    for class_index in list(annotation_mapping.values()):
        tp = np.sum((y_true[:, class_index] == 1) & (y_pred[:, class_index] == 1))
        fp = np.sum((y_true[:, class_index] == 0) & (y_pred[:, class_index] == 1))
        fn = np.sum((y_true[:, class_index] == 1) & (y_pred[:, class_index] == 0))

        precision = tp / (tp + fp) if tp + fp > 0 else 0
        recall = tp / (tp + fn) if tp + fn > 0 else 0
        if precision + recall > 0:
            f1_scores.append(2 * (precision * recall) / (precision + recall))
        else:
            f1_scores.append(0)
    return f1_scores


answer_columns = ['answerPrompt2_int', 'answerPrompt4_int', 'answerPrompt6_int', 'answerPrompt8_int', 'answerPrompt10_int', 'answerOLD_int']

results_df = pd.DataFrame(columns=['Run', 'Average F1_Macro_GPT35', 'Average F1_Macro_GPT4', 'Accuracy_GPT35', 'Accuracy_GPT4'])

mlb = MultiLabelBinarizer(classes=list(annotation_mapping.values()))

for system in answer_columns:
    if system != 'answerOLD_int':
        true_labels_gpt35 = mlb.fit_transform(gpt35['annotation_int'])
        predicted_labels_gpt35 = mlb.transform(gpt35[system])

        accuracy_gpt35 = custom_accuracy_per_class(true_labels_gpt35, predicted_labels_gpt35)
        f1_scores_gpt35 = custom_f1_score_per_class(true_labels_gpt35, predicted_labels_gpt35)

        average_f1_macro_gpt35 = sum(f1_scores_gpt35) / len(f1_scores_gpt35)
        average_accuracy_gpt35 = sum(accuracy_gpt35) / len(accuracy_gpt35)

        #print(f"Class-wise F1 scores for {system} on GPT-35:")
        #print(classification_report(true_labels_gpt35, predicted_labels_gpt35, target_names=annotation_mapping.keys()))

    true_labels_gpt4 = mlb.fit_transform(gpt4['annotation_int'])
    predicted_labels_gpt4 = mlb.transform(gpt4[system])

    accuracy_gpt4 = custom_accuracy_per_class(true_labels_gpt4, predicted_labels_gpt4)
    f1_scores_gpt4 = custom_f1_score_per_class(true_labels_gpt4, predicted_labels_gpt4)

    average_f1_macro_gpt4 = sum(f1_scores_gpt4) / len(f1_scores_gpt4)
    average_accuracy_gpt4 = sum(accuracy_gpt4) / len(accuracy_gpt4)

    new_row = pd.DataFrame([{
        'Run': system,
        'Average F1_Macro_GPT35': average_f1_macro_gpt35,
        'Average F1_Macro_GPT4': average_f1_macro_gpt4,
        'Accuracy_GPT35': accuracy_gpt35,
        'Accuracy_GPT4': accuracy_gpt4
    }])

    results_df = pd.concat([results_df, new_row], ignore_index=True)

print(results_df)

# Save results to a CSV file
results_df.to_csv('/content/drive/My Drive/Lavoro ChatGPT/dataset/Valeria/18/18_uc_results_perclass.csv', index=False)
```
